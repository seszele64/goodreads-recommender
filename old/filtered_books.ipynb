# filter data
- i want books with many reviews -> more than 1000
- with many shelves
- has similar books -> measure efficiency

```python


from pyspark.sql.functions import size

# filter out books with less than 100 ratings
df = df.filter(df.ratings_count >= 100)

# filter for at least 5 in popular_shelves
df = df.filter(size(df.popular_shelves) >= 5)

# has similar_books
df = df.filter(size(df.similar_books) > 0)

# with many shelves
df.show(5)

df.printSchema()

# print book with more than 1000 ratings
df.filter(df.ratings_count >= 1000).show(5)
```
