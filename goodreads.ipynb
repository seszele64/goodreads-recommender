```python
file_name = 'goodreads_books.json.gz'

with gzip.open(path + file_name) as f:
    # get column names
    first_line = f.readline()
    column_names = list(json.loads(first_line).keys())
    print(column_names)
    data = f.read(size=1000).decode('utf-8')
```

```python
# Dask
import dask.dataframe as dd
import pandas as pd
import random


# Initialize an empty list to store the data
data = []

# sample size
sample_size = 1000

# Open the Gzip-compressed JSON file and load random 1000 rows
with gzip.open(path+file_name, 'rb') as f:
    for line in f:
        # Generate a random number between 0 and the current line number
        if len(data) < sample_size:
            data.append(json.loads(line.decode('utf-8')))
        else:
            rand_num = random.randint(0, len(data))
            if rand_num < sample_size:
                data[rand_num] = json.loads(line.decode('utf-8'))

df = pd.DataFrame(data)
df
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
df=spark.read.json(path + 'goodreads_books.json.gz')

# filter out rows that have fewer than 10 ratings
df = df.filter(df.ratings_count > 10)
```

```python
from pyspark.sql.functions import avg, col, concat, desc, explode, lit, min, max, split, stddev, udf, isnan, when, count, countDistinct

# show how many rows are in the dataset
df.count()
```

```python
# calculate average rating by author and sort
avg_rating_df = df.groupBy('authors').agg(avg('average_rating')).sort(desc('avg(average_rating)'))
avg_rating_df.show()
```
